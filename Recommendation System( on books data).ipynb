```python
## Reshaping the data size using pivot
```

```python
df_NB["User.ID"].unique()
```

```python
df_NB2 = df_NB.pivot_table(index='User.ID', columns ="Book.Title" , values = "Book.Rating")
```

```python
df_NB2
```

```python
#Impute those NaNs with 0 values
```

```python
df_NB2.fillna(0, inplace=True)
```

```python
df_NB2
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
import numpy as np
```

```python
user_sim = 1 - pairwise_distances( df_NB2.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe  
user_sim_df = pd.DataFrame(user_sim)
```
